```python
#Funcion de distribucion de Laplace
loc = 0
scale = 2
pdf = np.exp(-abs(x-loc)/scale)/(2.*scale)
```

```python
#Funcion con kurtosis negativa Raised Cosine
x2 = np.linspace(-2,2,100) 
pdfRcos = (1.+np.cos(pi*(x2-loc)/scale))/(2.*scale)
```

```python
#Comparacion de Graficas
plt.plot(x,mlab.normpdf(x, loc, scale),label="Gaussian, \kappa = 0")
plt.plot(x, pdf,label="Laplace, \kappa = 3")
plt.plot(x2, pdfRcos,label="Rasied Cosine, \kappa = -0.6")
legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
       ncol=2, mode="expand", borderaxespad=0.)
plt.ylabel('p(x)',fontsize=18)
plt.xlabel('x',fontsize=18)
savefig('kurtosisL1.png', bbox_inches='tight',dpi=1200)
```
